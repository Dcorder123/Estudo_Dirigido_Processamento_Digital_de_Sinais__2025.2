# Resumo Teórico: Transformadas $Z$ e de Fourier

Este resumo aborda a Transformada $Z$, a Transformada de Fourier de Tempo Discreto (TFTD) e a Transformada de Fourier Discreta (TFD), destacando suas definições, relações e aplicações em sistemas lineares invariantes no tempo (LIT).
---



## 1. Transformada $Z$ e Região de Convergência (RDC)

A **Transformada $Z$** ($\mathcal{Z}$) de uma sequência $x[n]$, denotada por $X(z)$, é definida pela série de potências:
$$X(z) = \sum_{n=-\infty}^{\infty} x[n]z^{-n}$$

É a **generalização mais ampla** para a análise de sistemas de tempo discreto, convertendo equações de diferenças lineares em equações algébricas no **plano $z$**.

### Relação com a TFTD

A **TFTD**, $X(e^{j\omega})$, é obtida substituindo a variável complexa $z$ por $e^{j\omega}$:
$$X(e^{j\omega}) = X(z)|_{z=e^{j\omega}}$$
Isto corresponde a calcular $X(z)$ sobre o contorno de magnitude unitária no plano $z$, conhecido como **circunferência unitária** ($|z|=1$).

### Região de Convergência (RDC)

A **RDC** é o conjunto de valores de $z$ para os quais a série da Transformada $Z$ **converge**. Para Transformadas $Z$ racionais, ela é tipicamente um anel: $0 \le r_R < |z| < r_L \le \infty$.

* A transformada de Fourier de $x[n]$ converge em valor absoluto se, e somente se, a **RDC incluir a circunferência unitária**.
* A RDC **não pode conter quaisquer polos**.

### Relação Polos/Zeros e Estabilidade

Para uma função de sistema racional $X(z) = P(z)/Q(z)$:
* **Zeros** são as raízes do numerador, $P(z)$.
* **Polos** (para valores finitos de $z$) são as raízes do denominador, $Q(z)$.

| Conceito | Condição na Transformada $Z$ |
| :--- | :--- |
| **Causalidade** | Se $x[n]$ é **lateral direita** (causal se $n_1 \ge 0$), a RDC se estende para o exterior a partir do polo finito de **maior magnitude** até $z = \infty$. |
| **Estabilidade** | Exige que a **RDC inclua a circunferência unitária**. |
| **Estabilidade em Sistemas Causais (Racionais)** | Requer que **todos os polos** estejam no **interior do círculo unitário** ($|z| < 1$). |

A utilidade primária da $\mathcal{Z}$-Transformada é que para um sistema LIT, a $\mathcal{Z}$-Transformada da saída $y[n]$ é:
$$Y(z) = H(z)X(z)$$
onde $H(z)$ é a função de sistema.


## 2. Transformada de Fourier de Tempo Discreto (TFTD)

A **TFTD**, $X(e^{j\omega})$, é a representação no domínio da frequência para sinais e sistemas de tempo discreto.

### Natureza e Periodicidade

* É uma função de valores complexos da variável **contínua** $\omega$ (frequência).
* É **periódica** em $\omega$ com período **$2\pi$**.

### Propriedades Essenciais

| Propriedade | Relação no Domínio do Tempo | Relação no Domínio da Frequência |
| :--- | :--- | :--- |
| **Linearidade** | $ax_1[n] + bx_2[n]$ | $aX_1(e^{j\omega}) + bX_2(e^{j\omega})$ |
| **Convolução** | $x[n] * h[n]$ | $X(e^{j\omega})H(e^{j\omega})$ |
| **Modulação** | $x[n]w[n]$ | $\frac{1}{2\pi} \int_{-\pi}^{\pi} X(e^{j\theta}) W(e^{j(\omega-\theta)})d\theta$ (**Convolução Periódica**) |

Para sistemas LIT, a saída no domínio da frequência é $Y(e^{j\omega}) = H(e^{j\omega})X(e^{j\omega})$.

## 3. Transformada de Fourier Discreta (TFD) e FFT

A **Transformada de Fourier Discreta (TFD)** é uma representação de Fourier para **sequências de duração finita**.

### Relação com a TFTD (Amostragem Espectral)

A TFD, $X[k]$, é uma sequência **discreta** que corresponde a **amostras em frequência** da TFTD do sinal, igualmente espaçadas:
$$X[k] = X(e^{j\omega})|_{\omega = 2\pi k/N}, \quad k=0, 1, \dots, N-1$$
onde $N$ é o comprimento da TFD.

### Propriedades da TFD

* **Periodicidade:** $X[k]$ é periodicamente estendida e representa uma sequência periódica $x̃[n]$ no domínio do tempo.
* **Linearidade:** A TFD é linear.
* **Convolução (Circular):** O produto de duas TFDs de $N$ pontos ($X_1[k]X_2[k]$) corresponde a uma **convolução circular** no domínio do tempo: $x_{1}[n] \circledast_{N} x_{2}[n]$.
    * A convolução circular é a convolução linear seguida pelo aliasing no tempo (sobreposição de réplicas periódicas).

### Cálculo via FFT

O termo **Transformada de Fourier Rápida (FFT)** refere-se a uma família de algoritmos para o **cálculo eficiente** da TFD de $N$ pontos.

[Image of butterfly diagram for 8-point FFT]


* **Eficiência:** O cálculo direto da TFD requer $O(N^2)$ operações. Os algoritmos de FFT (e.g., dizimação no tempo/frequência) requerem $O(N \log_2 N)$ operações, tornando a análise espectral prática.

## 4. Diferenças e Perspectiva Complementar

| Característica | TFTD ($X(e^{j\omega})$) | TFD ($X[k]$) |
| :--- | :--- | :--- |
| **Variável de Frequência** | **Contínua** ($\omega$) | **Discreta** ($k$) |
| **Cálculo** | Representação ideal, teórica (soma infinita/integral inversa) | Calculada por soma finita (FFT) |
| **Convolução** | Multiplicação $\leftrightarrow$ **Convolução Linear** | Multiplicação $\leftrightarrow$ **Convolução Circular** |
| **Natureza** | Representação ideal | Amostragem discreta da TFTD |

### Perspectiva Complementar

| Transformada | Foco Principal | Aplicações Chave |
| :--- | :--- | :--- |
| **$\mathcal{Z}$ (Análise Estrutural)** | **Generalização mais ampla**. Localização de **Polos/Zeros** e a **RDC**. | Análise de **estabilidade** e **causalidade** dos sistemas LIT. |
| **TFTD (Análise Teórica de Frequência)** | Representação espectral **contínua** ideal. | Entendimento do **comportamento de frequência** dos sistemas (Resposta em Frequência $H(e^{j\omega})$). |
| **TFD (Análise Computacional Prática)** | **Ferramenta computável** (via FFT). | **Análise espectral** eficiente, implementação de **convoluções lineares** (usando convolução circular controlada). |

### 1\. Configuração Inicial e Importação de Bibliotecas

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.fft import fft, fftshift, fftfreq
from scipy.signal import windows, freqz, tf2zpk

# Configurações de plotagem para melhor visualização
plt.rcParams['figure.figsize'] = [12, 6]
plt.rcParams['axes.grid'] = True
```


### a) e b) Geração do Sinal, Cálculo da DFT e Plotagem (Magnitude e Fase)

Aqui geramos o sinal $x[n] = \sin(2\pi 0.1n) + 0.5 \sin(2\pi 0.4n)$. Note que as frequências 0.1 e 0.4 são frequências normalizadas ($f/F_s$), onde 0.5 seria a frequência de Nyquist.


```python
# --- a) Geração do Sinal ---
N = 64  # Número de amostras
n = np.arange(N)

# Frequências normalizadas (ciclos/amostra)
f1 = 0.1
f2 = 0.4

# x[n] = sin(2π * 0.1 * n) + 0.5 * sin(2π * 0.4 * n)
x = np.sin(2 * np.pi * f1 * n) + 0.5 * np.sin(2 * np.pi * f2 * n)

# --- b) Cálculo da DFT ---
X = fft(x)

# Frequências correspondentes (de 0 a 1 normalizado ou -0.5 a 0.5)
# Vamos usar fftshift para centralizar o espectro em 0
freqs = fftshift(fftfreq(N))
X_shifted = fftshift(X)

# Magnitude e Fase
magnitude = np.abs(X_shifted)
fase = np.angle(X_shifted)

# Limpeza da fase: zerar fase onde a magnitude é muito pequena (ruído numérico)
fase[magnitude < 1e-5] = 0

# --- Plotagem ---
plt.figure(figsize=(14, 6))

# Plot Magnitude
plt.subplot(1, 2, 1)
plt.stem(freqs, magnitude)
plt.title(f'Magnitude da DFT (N={N})')
plt.xlabel('Frequência Normalizada (ciclos/amostra)')
plt.ylabel('|X[k]|')
plt.grid(True)

# Plot Fase
plt.subplot(1, 2, 2)
plt.stem(freqs, fase)
plt.title('Fase da DFT')
plt.xlabel('Frequência Normalizada')
plt.ylabel('Fase (radianos)')
plt.grid(True)

plt.tight_layout()
plt.show()
```

### c) Aplicação de Janelas (Retangular, Hamming, Blackman)

Para ver o efeito das janelas, vamos mudar ligeiramente a frequência para uma que não seja inteira em relação a N (para forçar o *spectral leakage*) e visualizar em escala logarítmica (dB).


```python
# Vamos usar um sinal que causa vazamento (frequência não binada exatamente)
N_win = 64
n_win = np.arange(N_win)
# Frequência 0.12 não é um múltiplo inteiro exato da resolução base se N=64
x_leak = np.sin(2 * np.pi * 0.12 * n_win)

# Criar janelas
win_rect = np.ones(N_win)             # Retangular
win_hamm = np.hamming(N_win)          # Hamming
win_black = np.blackman(N_win)        # Blackman

# Aplicar janelas
x_rect = x_leak * win_rect
x_hamm = x_leak * win_hamm
x_black = x_leak * win_black

# Calcular DFT (com zero-padding para suavizar visualmente o espectro - DTFT aproximada)
N_fft = 512
X_rect = fftshift(fft(x_rect, N_fft))
X_hamm = fftshift(fft(x_hamm, N_fft))
X_black = fftshift(fft(x_black, N_fft))

freqs_win = fftshift(fftfreq(N_fft))

# Função para converter para dB
def db(x):
    return 20 * np.log10(np.abs(x) / np.max(np.abs(x)) + 1e-10) # normalizado

# Plotagem comparativa
plt.figure(figsize=(12, 6))
plt.plot(freqs_win, db(X_rect), label='Retangular (Boxcar)', linewidth=1.5)
plt.plot(freqs_win, db(X_hamm), label='Hamming', linewidth=1.5)
plt.plot(freqs_win, db(X_black), label='Blackman', linewidth=1.5)

plt.title('Comparação de Janelas: Efeito no Vazamento Espectral (Leakage)')
plt.xlabel('Frequência Normalizada')
plt.ylabel('Magnitude Normalizada (dB)')
plt.ylim([-100, 5])
plt.legend()
plt.grid(True)
plt.show()
```

**Observação Teórica (para o relatório):**

  * **Retangular:** Lóbulo principal estreito (boa resolução), mas lóbulos laterais altos (muito *leakage*/vazamento).
  * **Blackman:** Lóbulo principal largo (pior resolução), mas lóbulos laterais muito baixos (pouco vazamento).
  * **Hamming:** Um meio termo.


### d) Efeito do tamanho da DFT (N) na Resolução e Leakage

Aqui vamos criar dois sinais com frequências muito próximas e tentar separá-los usando diferentes tamanhos de N.

```python
# Sinal com duas frequências próximas: 0.2 e 0.22
def generate_signal(N_samples):
    n_local = np.arange(N_samples)
    return np.sin(2*np.pi*0.2*n_local) + np.sin(2*np.pi*0.22*n_local)

# Caso 1: N pequeno (Baixa resolução)
N1 = 32
x1 = generate_signal(N1)
X1 = fftshift(fft(x1))
f1_axis = fftshift(fftfreq(N1))

# Caso 2: N grande (Alta resolução)
N2 = 256
x2 = generate_signal(N2)
X2 = fftshift(fft(x2))
f2_axis = fftshift(fftfreq(N2))

# Plotagem
plt.figure(figsize=(14, 5))

plt.subplot(1, 2, 1)
plt.stem(f1_axis, np.abs(X1))
plt.title(f'Baixa Resolução (N={N1})\nImpossível distinguir 0.2 e 0.22')
plt.xlabel('Freq')
plt.ylabel('|X|')
plt.grid(True)

plt.subplot(1, 2, 2)
plt.stem(f2_axis, np.abs(X2))
plt.title(f'Alta Resolução (N={N2})\nPicos separados em 0.2 e 0.22')
plt.xlabel('Freq')
plt.ylabel('|X|')
plt.xlim([0.1, 0.3]) # Zoom na área de interesse
plt.grid(True)

plt.show()
```

### e) Análise de Sistema LTI: Polos/Zeros e Resposta em Frequência

Vamos analisar um filtro simples (Média Móvel) para ver a relação entre os zeros e a magnitude.

$H(z) = \frac{1}{2}(1 - z^{-1})$ (Filtro Passa-Alta simples)

```python
def plot_zplane(b, a):
    """Função auxiliar para plotar o plano Z com polos e zeros"""
    # Calcular polos e zeros
    z, p, k = tf2zpk(b, a)

    # Criar figura e círculo unitário
    plt.figure(figsize=(12, 5))
    ax1 = plt.subplot(1, 2, 1)
    unit_circle = plt.Circle((0,0), 1, color='black', fill=False, linestyle='--')
    ax1.add_patch(unit_circle)

    # Plotar polos (x) e zeros (o)
    plt.plot(np.real(z), np.imag(z), 'go', markersize=10, label='Zeros')
    plt.plot(np.real(p), np.imag(p), 'rx', markersize=10, label='Polos')

    plt.xlim([-1.5, 1.5])
    plt.ylim([-1.5, 1.5])
    plt.grid(True)
    plt.axhline(0, color='black', linewidth=0.5)
    plt.axvline(0, color='black', linewidth=0.5)
    plt.title('Mapa de Polos e Zeros')
    plt.legend()
    plt.xlabel('Real')
    plt.ylabel('Imaginário')
    return ax1

# Definição do Sistema (Exemplo: Filtro FIR simples)
# H(z) = 1 + z^-1 + z^-2 (Média móvel de 3 pontos, passa-baixa)
b = [1, 1, 1]
a = [1]       # FIR, denominador é 1

# 1. Plotar Polos e Zeros
plot_zplane(b, a)

# 2. Plotar Resposta em Frequência
w, h = freqz(b, a, worN=512)

plt.subplot(1, 2, 2)
plt.plot(w / np.pi, np.abs(h))
plt.title('Resposta em Frequência |H(e^jw)|')

# --- CORREÇÃO AQUI: Adicionado o 'r' antes da string ---
plt.xlabel(r'Frequência Normalizada (x $\pi$ rad/amostra)')

plt.ylabel('Magnitude')
plt.grid(True)

plt.tight_layout()
plt.show()
```

**O que observar no item (e):**

1.  Os **Zeros** "puxam" a resposta em frequência para baixo (magnitude tende a zero quando a frequência angular $\omega$ se aproxima do ângulo do zero no círculo unitário).
2.  Os **Polos** "empurram" a resposta para cima (picos de ressonância).
3.  Neste exemplo (média móvel), os zeros estão em frequências mais altas, atenuando-as (comportamento Passa-Baixa).
