# **Parte 1 – Fundamentos, Sistemas Discretos e Convolução**



## **1 - Resumo Teórico**

### **1. Fundamentos dos Sistemas de Tempo Discreto**
Os sistemas de tempo discreto são transformações matemáticas que relacionam uma **sequência de entrada** $x[n]$ a uma **sequência de saída** $y[n]$, descritas genericamente por:

$$
y[n] = T\{x[n]\}
$$

onde $T\{\cdot\}$ representa o operador ou regra do sistema. Cada amostra de saída pode depender de uma ou várias amostras de entrada, passadas, presentes ou futuras.




### **2. Classificação dos Sistemas Discretos**

#### **2.1. Sistemas com e sem Memória**
- **Sem memória:** a saída em um instante $n$ depende apenas do valor atual da entrada $x[n]$.  
  Exemplo: $y[n] = 3x[n]$
- **Com memória:** a saída depende de amostras passadas e/ou futuras da entrada.  
  Exemplo: $y[n] = x[n] - x[n-1]$

#### **2.2. Sistemas Lineares e Não Lineares**
Um sistema é **linear** se satisfaz o **princípio da superposição**:

$$
T\{\alpha x_1[n] + \beta x_2[n]\} = \alpha\, T\{x_1[n]\} + \beta\, T\{x_2[n]\}
$$

Caso contrário, é **não linear**.

#### **2.3. Sistemas Invariantes e Variantes no Tempo**
Um sistema é **invariante no tempo (LTI)** se o deslocamento da entrada gera o mesmo deslocamento na saída:

$$
T\{x[n-n_0]\} = y[n-n_0]
$$

Caso contrário, é **variante no tempo**.

#### **2.4. Sistemas Causais e Não Causais**
- **Causais:** a saída $y[n_0]$ depende apenas de valores de entrada $x[n]$ para $n \le n_0$.  
  Exemplo: $y[n] = x[n] - x[n-1]$
- **Não causais:** a saída depende de valores futuros da entrada.  
  Exemplo: $y[n] = x[n+1] - x[n]$

#### **2.5. Estabilidade BIBO (Entrada Limitada → Saída Limitada)**
Um sistema é **estável** se toda entrada limitada produzir uma saída limitada:

$$
|x[n]| \le B_x < \infty \quad\Rightarrow\quad |y[n]| \le B_y < \infty
$$

A instabilidade ocorre quando uma entrada limitada gera uma saída ilimitada.

#### **2.6. Sistemas Determinísticos e Aleatórios**
- **Determinístico:** a saída é previsível a partir da entrada.  
- **Aleatório:** envolve variáveis estatísticas ou ruído; o comportamento é probabilístico.

#### **2.7. Sistemas Inversíveis e Não Inversíveis**
Um sistema é **inversível** se existir outro sistema $T^{-1}$ tal que:

$$
T^{-1}\{T\{x[n]\}\} = x[n]
$$

Caso contrário, é **não inversível**.





### **3. Propriedades Formais dos Sistemas**

| **Propriedade** | **Condição Matemática** | **Descrição** |
|-----------------|--------------------------|----------------|
| **Linearidade** | $T\{\alpha x_1 + \beta x_2\} = \alpha T\{x_1\} + \beta T\{x_2\}$ | Satisfaz superposição |
| **Invariância no tempo** | $T\{x[n-n_0]\} = y[n-n_0]$ | Resposta não muda no tempo |
| **Causalidade** | $y[n_0]$ depende de $x[n]$ apenas para $n \le n_0$ | Não antecipa entradas |
| **Estabilidade (BIBO)** | Entrada limitada → saída limitada | Saída não diverge |
| **Memória** | Depende de $x[n]$, $x[n-k]$ ou $x[n+k]$ | Sistema armazena histórico |





### **4. Resposta ao Impulso e Representação por Convolução**
Um sistema **linear e invariante no tempo (LTI)** é completamente caracterizado por sua **resposta ao impulso** $h[n]$.  
A saída para qualquer entrada $x[n]$ é obtida pela **soma de convolução**:

$$
y[n] = \sum_{k=-\infty}^{\infty} x[k]\,h[n-k]
$$

ou, de forma compacta:

$$
y[n] = x[n] * h[n]
$$

**Interpretação:** Cada amostra $x[k]$ da entrada multiplica uma cópia deslocada de $h[n]$, e a saída é a soma ponderada dessas versões deslocadas.





### **5. Propriedades da Convolução**
1. **Comutatividade:** $x[n] * h[n] = h[n] * x[n]$
2. **Associatividade:** $x[n] * (h[n] * g[n]) = (x[n] * h[n]) * g[n]$
3. **Distributividade:** $x[n] * (h_1[n] + h_2[n]) = x[n] * h_1[n] + x[n] * h_2[n]$
4. **Duração:** Se $x[n]$ possui $N_x$ amostras e $h[n]$ possui $N_h$, então $N_y = N_x + N_h - 1$.





### **6. Interpretação Física e Computacional**
A convolução descreve a **superposição temporal das respostas ao impulso**.  
Em termos computacionais, envolve:
1. Refletir $h[k]$ → $h[-k]$;
2. Deslocar $h[-k]$ por $n$ → $h[n-k]$;
3. Multiplicar ponto a ponto $x[k]h[n-k]$;
4. Somar os produtos → $y[n]$.

Esse processo representa o **efeito acumulado da entrada sobre o sistema**, onde $h[n]$ define sua memória e comportamento dinâmico.


### **7. Considerações Finais**
- Sistemas LTI são a base do processamento digital de sinais.  
- A resposta ao impulso contém toda a informação do sistema.  
- A estabilidade e causalidade podem ser analisadas a partir de $h[n]$.  
- A convolução é a principal ferramenta para simular e implementar sistemas discretos.

#**2 - Simulação em Python**

### Definição dos sitemas:

```python
import numpy as np

# Função degrau unitário
def u(n):
    return np.where(n >= 0, 1, 0)

# Sistema T1: Diferenciador discreto (LTI)
# T1{x[n]} = x[n] - x[n-1]
def T1(x):
    y = np.zeros_like(x)
    y[0] = x[0]            # assume x[-1] = 0
    y[1:] = x[1:] - x[:-1]
    return y

# Sistema T2: Variante no tempo (não LTI)
# T2{x[n]} = (0.8)^n * x[n]
def T2(x, n):
    return (0.8 ** n) * x

# Sinais de teste
n = np.arange(0, 21)
x1 = (n == 3).astype(float)                          # impulso deslocado
x2 = np.where((n >= 5) & (n <= 10), 1.0, 0.0)        # pulso retangular
```

### Teste de prioridades:

```python
alpha, beta = 2.0, -0.5

# TESTES PARA T1 (LTI)
# Linearidade
LHS = T1(alpha*x1 + beta*x2)
RHS = alpha*T1(x1) + beta*T1(x2)
print("T1 é linear:", np.allclose(LHS, RHS))

# Invariância no tempo
n0 = 4
x_shift = np.roll(x1, n0)
y = T1(x1)
y_shift = T1(x_shift)
print("T1 é invariante no tempo:", np.allclose(y_shift, np.roll(y, n0)))

# TESTES PARA T2 (não LTI)
LHS2 = T2(alpha*x1 + beta*x2, n)
RHS2 = alpha*T2(x1, n) + beta*T2(x2, n)
print("T2 é linear:", np.allclose(LHS2, RHS2))

y2 = T2(x1, n)
y2_shift = T2(x_shift, n)
print("T2 é invariante no tempo:", np.allclose(y2_shift, np.roll(y2, n0)))
```

### Convolução:

```python
import matplotlib.pyplot as plt

# h[n] = (0.8)^n u[n]
h = (0.8 ** n) * u(n)

# x[n] = u[n]
x = u(n)

# Convolução
y_conv = np.convolve(x, h)
n_y = np.arange(len(y_conv))

# Gráficos
plt.figure(figsize=(6,3))
plt.stem(n, x)
plt.title("Entrada x[n] = u[n]")
plt.grid(True)
plt.show()

plt.figure(figsize=(6,3))
plt.stem(n, h)
plt.title("Resposta ao impulso h[n] = (0.8)^n u[n]")
plt.grid(True)
plt.show()

plt.figure(figsize=(6,3))
plt.stem(n_y, y_conv)
plt.title("Saída y[n] = x[n] * h[n] (convolução)")
plt.grid(True)
plt.show()

# Comprimentos
print("len(x) =", len(x))
print("len(h) =", len(h))
print("len(y) =", len(y_conv), "  (deve ser len(x)+len(h)-1)")
```

 - Causalidade: sim (depende de $x[n]$ e $x[n-1]$)

 - Estabilidade BIBO: estável

**Conclusão:** T1 é LTI, causal, estável e com memória.

## **Sistema T2 — Sistema variante no tempo**

 - Causalidade: sim (depende apenas de $x[n]$)

**Conclusão:** T2 é linear, mas variante no tempo, causal, estável e sem memória.

#**4 - Discussão**


## **Relação entre as Propriedades Teóricas e o Comportamento de T1 e T2**

### **Sistema T1 — Diferenciador discreto**
O sistema T1 é definido por:

$$
T_1\{x[n]\} = x[n] - x[n-1]
$$

E apresenta, na prática, todas as propriedades esperadas de um sistema **LTI**:

- Linearidade: satisfeita  
- Invariância no tempo: satisfeita  
- Causalidade: depende apenas de $x[n]$ e $x[n-1]$  
- Estabilidade BIBO: estável  
- Memória: possui memória  

**Interpretação:** T1 atua como um diferenciador discreto, realçando variações do sinal e suprimindo regiões constantes.


### **Sistema T2 — Variante no tempo**
O sistema T2 é definido por:

$$
T_2\{x[n]\} = (0.8)^n x[n]
$$

E apresenta:

- Linearidade: satisfeita  
- Invariância no tempo: **não satisfeita**  
- Causalidade: sim  
- Estabilidade BIBO: estável  
- Memória: sem memória  

**Interpretação:** T2 atenua progressivamente o sinal ao longo do tempo, efeito típico de sistemas variantes no tempo.

## **Relação entre Resposta ao Impulso e Convolução**

A resposta ao impulso utilizada foi:

$$
h[n] = (0.8)^n u[n]
$$

O sistema é:

- Causal (h[n]=0 para n<0)  
- Estável (PG convergente)  
- Com memória curta (decaimento exponencial)  


## ** Interpretação da Saída da Convolução**

A convolução resultante é:

$$
\displaystyle
y[n] = \sum_{k=0}^{n} (0.8)^{\,n-k}
$$

No regime permanente:

$$
\displaystyle
y[\infty] = \sum_{k=0}^{\infty} (0.8)^k = \frac{1}{1-0.8} = 5
$$

Observações:

- Crescimento inicial devido ao acúmulo de cópias de $h[n]$  
- Comportamento suave devido ao decaimento exponencial  
- Convergência ao valor final 5  


## **Conclusão da Discussão**

- As simulações confirmam plenamente as propriedades teóricas dos sistemas discretos.  
- Somente T1 (LTI) pode ser representado por convolução.  
- A forma da saída reflete diretamente a superposição de respostas ao impulso.  
- O decaimento exponencial de $h[n]$ governa o comportamento transiente e estacionário.
